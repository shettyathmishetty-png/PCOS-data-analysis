```python
sns.countplot(x='PCOS_Diagnosis',data=df)
plt.show()
```

#### Observation:Distribution of PCOS Diagonis  

The datasetis imbalanced with significantly higher number of individual without PCOS(80%)
compared to those with PCOS (20%).This indicates That PCOS cases are less frequent in the dataset

```python
sns.boxplot(x='PCOS_Diagnosis',y='BMI',data=df)
plt.show()
```

### Observation: BMI Distribution (Box Plot)

The box plot shows that individuals with PCOS have a higher median BMI compared to those without PCOS. The BMI values for the PCOS group are more concentrated within a narrower range, while the non-PCOS group shows greater variability. Although both groups have a similar overall range of BMI values, the distribution for the PCOS group is shifted towards higher values, indicating that individuals with PCOS tend to have higher BMI.

```python
sns.boxplot(x='PCOS_Diagnosis',y='Testosterone_Level(ng/dL)',data=df)
plt.show()
```

## Observation: Testosterone Distribution and PCOS

The box plot shows that individuals with PCOS have a higher median testosterone level compared to those without PCOS, and the distribution is shifted towards higher values, indicating that testosterone levels tend to be higher in individuals diagnosed with PCOS.

### Additional Observation

The non-PCOS group shows greater variability in testosterone levels, as indicated by a wider spread in the box plot, while the PCOS group has a more concentrated distribution.
